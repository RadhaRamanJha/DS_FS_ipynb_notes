## Bar Charts :-

### Bar Charts can be used 
1. When we want to show how some quantity varies among some discrete set of items

```python
movies = ["Gully Boy", "Devdas", "Black", "Dum", "Omkara"]
num_fimfare = [5, 11, 3, 8, 10]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_fimfare)

plt.title("Hyped Movies")     # add a title
plt.ylabel("# of Fimfare Awards")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)

plt.show()
```

### Bar Charts can be used 

2. Bar charts can also be a good choice for plotting histograms of bucketed numeric values, in order to visually explore how the values are distributed:

```python
from collections import Counter
grades = [25,25,45,46,65,75,80,90,95,22,93,68,78,98,86,46,47,89,99]
# Bucket grades by decile, but put 100 in with the 90s (i.e. truncates the numbers to nearest lower 10)
histogram = Counter(min(grade//10*10,90) for grade in grades)
plt.bar(x = [x+5 for x in histogram.keys()],                  # Shift bars right by 5
        height = histogram.values(),                          # Give each bar its correct height
        width = 10,                                           # Give each bar a width of 10
        edgecolor = (0,0,0))                                  # Black edge for each bar
plt.axis([-5,105,0,6])                                        # Set axis limit [xmin,xmax,ymin,ymax]

plt.xticks([10*i for i in range(11)])                         # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Marks")
plt.show()
```

### Notes for the above plot :- 
1. for argument plt.bar :-

   a. We use bar width so that it fills the entire decile range
   
   b. Centers of all bars are shifted 5 units so that it occupies the correct range
   
   c. Black edge is specified mentioning " edgecolor = (0,0,0) " so that bars are distinctly visible
   

2. the call to plt.axis indicates :- 

   a. We want our x-axis start from -5 and end at 105
   
   b. We want our y-axis start from 0 and end at 6
   

3. call to plt.xticks puts x-axis labels at 0, 10, 20, â€¦, 100.

## While creating bar plot, it is good to start y-axis from 0 because it can be misleading

```python
times_said = [300, 305]
years = [2023, 2024]

plt.bar(years, times_said, 0.8)
plt.xticks(years)
plt.ylabel("# of times I read 'data science'")

plt.ticklabel_format(useOffset=False)
# this argument is for formating the tick labels especially useful while plotting very large or very small number

# misleading y-axis only shows the part above 299
plt.axis([2022.5, 2024.5, 299, 306])
plt.title("Look at the 'Huge' Increase!")
plt.show()
```

```python
times_said = [300, 305]
years = [2023, 2024]

plt.bar(years, times_said, 0.8)
plt.xticks(years)
plt.ylabel("# of times I read 'data science'")

plt.ticklabel_format(useOffset=False)     
# this argument is for formating the tick labels especially useful while plotting very large or very small number

# misleading y-axis only shows the part above 500
plt.axis([2022.5, 2024.5, 0, 350])
plt.title("Difference not Huge anymore!")
plt.show()
```

## Line Charts :- 

1. Line charts made using plt.plot()

2. They are good choice for showing trend (a general direction in which something is developing or changing)

```python
variance = [1,2,4,8,16,32,64,128,256,512]
bias_squared = [512,256,128,64,32,16,8,4,2,1]
total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i,_ in enumerate(variance)]                           # to plot x-axis

# It is possible to make multiple call to plt.plot() to display multiple lines on same chart
plt.plot(xs,variance,'g-.',label = 'variance') # greeen dot-dashed line
plt.plot(xs,bias_squared,'r-',label = 'bias^2') # red solid line
plt.plot(xs,total_error,'b:',label = 'total error') # blue dotted line

# because we have already added label to each series we can get legends for free
plt.legend(loc = 9)
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

## Scatterplots :- 

### A scatterplot is the right choice for visualizing the relationship between two paired sets of data.

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
        xy=(friend_count, minute_count), # Put the label with its point
        xytext=(5, -5),                  # but slightly offset
        textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on the site")
plt.show()
```

#### But if we scatter comparable variables, we might get a misleading picture if we let matplotlib choose the scale

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.show()
```

## But if we include plt.axis("equal") the plot will show the plot more accurately shows that most of the variation occurs on test 2.

```python
test_1_grades = [ 99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Are Comparable now")
plt.xlabel("test 1 grade")
plt.axis("equal")
plt.ylabel("test 2 grade")
plt.show()
```
